```python
#importing data into a data frame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/Users/jasrajchhokar/Desktop/Assignment 1 Data Science/ChinaMainLandDataset.xlsx', sheet_name = "Total") # df means dataframe

# print whole sheet data
print(df)

print("Line graph:  ") 
plt.plot(df["Day"], df["confirmed"]) 
plt.title('Daily China Mainland covid confirmed cases (22/02/20 - 1/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Confirmed Cases')
plt.show()

print("Line graph:  ") 
plt.plot(df["Day"], df["deaths"]) 
plt.title('Daily China Mainland covid deaths (22/02/20 - 1/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Deaths')
plt.show()

print("Line graph:  ") 
plt.plot(df["Day"], df["recovered"]) 
plt.title('Daily China Mainland covid recovered cases (22/02/20 - 1/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Recovered Cases')
plt.show()
```
