```python
#importing data into a data frame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('AfghanistanDataset.xlsx', sheet_name = "January Total") # df means dataframe

# print whole sheet data
print(df)

print("Line graph:  ") 
plt.plot(df["Days"], df["confirmed"]) 
plt.title('Daily Afghanistan covid confirmed cases (1/01/2021-31/01/2021)')
plt.xlabel('Day')
plt.ylabel('Number of Confirmed Cases')
plt.show()

print("Line graph:  ") 
plt.plot(df["Days"], df["deaths"]) 
plt.title('Daily Afghanistan covid deaths (1/01/2021-31/01/2021)')
plt.xlabel('Day')
plt.ylabel('Number of Deaths')
plt.show()

print("Line graph:  ") 
plt.plot(df["Days"], df["recovered"]) 
plt.title('Daily Afghanistan covid recovered cases (1/01/2021-31/01/2021)')
plt.xlabel('Day')
plt.ylabel('Number of Recovered Cases')
plt.show()

```
