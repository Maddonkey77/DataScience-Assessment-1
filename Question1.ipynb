```python
#importing data into a data frame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('cases.xlsx', sheet_name = "Deaths2023") # df means dataframe

# print whole sheet data
print(df)
```

```python
# deaths
print("Line graph:  ") 
plt.plot(df["Day"], df["Total UK"]) 
plt.title('Daily total UK covid death cases (01-31 January 2023)')
plt.xlabel('Day')
plt.ylabel('Number of Cases')
plt.show()
```

```python
confirmeddf = pd.read_excel('cases.xlsx', sheet_name = "Confirmed2023") # df means dataframe

# print whole sheet data
print(confirmeddf)
```

```python
#confirmed cases
print("Line Plot:  ") 
plt.plot(confirmeddf["Day"], confirmeddf["Total UK"]) 
plt.title('Daily Total UK covid confirmed cases (01-31 January 2023)')
plt.xlabel('Day')
plt.ylabel('Number of Cases')
plt.show()
```
