```python
#importing data into a data frame
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('UK&US2020Dataset.xlsx', sheet_name = "UKTotal2020") # df means dataframe

# print whole sheet data
print(df)

print("Line graph:  ") 
plt.plot(df["Days"], df["confirmed"]) 
plt.title('Daily UK covid confirmed cases (22/01/2020--01/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Confirmed Cases')
plt.show()

print("Line graph:  ") 
plt.plot(df["Days"], df["deaths"]) 
plt.title('Daily UK covid deaths (22/01/2020-01/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Deaths')
plt.show()

print("Line graph:  ") 
plt.plot(df["Days"], df["recovered"]) 
plt.title('Daily UK covid recovered cases (22/01/2020-01/03/2020)')
plt.xlabel('Day')
plt.ylabel('Number of Recovered Cases')
plt.show()
```
