```python
#Q3
import pandas as pd
import matplotlib.pyplot as plt
q3condf = pd.read_excel('UKcasestotal2.xlsx', sheet_name = "Confirmed2022") # df means dataframe

# print whole sheet data
print(q3condf)
```

```python
print("2022 Line graph:  ") 
plt.plot(q3condf["Day"], q3condf["Total UK"]) 
plt.title('Daily total UK covid confirmed cases (01-31 January 2022)')
plt.xlabel('Day')
plt.ylabel('Number of Cases')
plt.show()
```

```python
q3deathdf = pd.read_excel('UKcasestotal2.xlsx', sheet_name = "Deaths2022") # df means dataframe

# print whole sheet data
print(q3deathdf)
```

```python
print("2022 Line graph:  ") 
plt.plot(q3deathdf["Date"], q3deathdf["Total UK"]) 
plt.title('Daily total UK covid death cases (01-31 January 2022)')
plt.xlabel('Day')
plt.ylabel('Number of Cases')
plt.show()
```
